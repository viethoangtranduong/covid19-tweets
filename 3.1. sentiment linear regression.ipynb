```
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
import pandas as pd
```

```
target = np.load('target.npy')/4
sent_input = np.load('sentiment_input.npy')

print(sent_input.shape)
print(target.shape)
```

### Linear Regression

```
lm = LinearRegression()

X = sent_input
y = target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=7)

lm.fit(X_train,y_train)
```

```
predicts = lm.predict(X_test)

print(y_test[:100])

print("""
        Mean Squared Error: {}
        R2 Score: {}
        Mean Absolute Error: {}
     """.format(
        metrics.mean_squared_error(y_test, predicts),
        metrics.r2_score(y_test,predicts),
        metrics.mean_absolute_error(y_test,predicts)
        ))
```

```
error_sent = pd.DataFrame({
        'Actual Values': np.array(y_test).flatten(),
        'Predicted Values': predicts.flatten()})

error_sent
```
